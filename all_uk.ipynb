```python
def get_area_data(df, area_code):
    
    df = df.set_index('Area code')
    if area_code in df.index:
        df_area = df.loc[area_code]
        df_area_row = df_area[:1]
        df_area_row = df_area.loc[:,['Area name', 'Cumulative lab-confirmed cases']]
        df_area_row = df_area_row[:1]
        df_area_row.reset_index(inplace=True)
        return df_area_row.iloc[0,:]
```

```python
df_area = get_area_data(df_data, "E06000030")
df_area
```

```python
def get_all_eng_data(df, code_list):
    
    # Initialise main dataframe
    col_names = ['Area code', 'Area name', 'Cumulative lab-confirmed cases']
    my_df = pd.DataFrame(columns = col_names)
    
    for code in code_list:
        
        my_df.loc[len(my_df)] = get_area_data(df_data, code)
        
    my_df = my_df.set_index('Area code')
    my_df = my_df.dropna()
    return my_df
```

```python
df_england_data = get_all_eng_data(df_data, code_list)
```

```python
def get_area_data_chart(df, area_code):
    
    df = df.set_index('Area code')
    if area_code in df.index:
        df_area = df.loc[area_code]
        df_area = df_area.loc[:,['Area name', 'Specimen date', 'Cumulative lab-confirmed cases']]
        df_area = df_area.drop_duplicates()
        df_area = df_area.drop(columns=['Area name'])
        df_area = df_area.rename(columns={'Specimen date':'date', "Cumulative lab-confirmed cases":'value'})
        #df_area_row = df_area_row[:1]
        #df_area_row.reset_index(inplace=True)
        return df_area[::-1]
    else:
        return pd.DataFrame.empty
```
